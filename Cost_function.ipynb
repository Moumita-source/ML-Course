```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import numpy library - It is used for scientific computing
### Import matplotlib and seaborn library - Both of these will be used for visualization

```python
## Here, we have a data, which includes area of the house in sq feet and price of the house in dollars
```

```python
X_train = np.array([2600, 3000, 3200, 3600, 4000])
print(f"Training set of area of house in sq feet : {X_train}")
print(f"Length of the total training data set : {X_train.shape[0]}")
```

```python
y_train = np.array([550000, 565000, 610000, 680000, 725000])
print(f"Labels of our training data (price in $): {y_train}")
print(f"Length of the labels for the training data : {y_train.shape[0]}")
```

### Lets visualize our training data 

```python
fig, ax = plt.subplots()
ax.set_title("Area of the house Vs Price of the house")
sns.lineplot(x = X_train, y = y_train)
ax.set_xlabel("Area of house (sq ft)")
ax.set_ylabel("Price of house (dollars)")
plt.show()
```

```python
## We mostly see an increasing trend as with the increase in area of the house , price of the house is also increasing
```

### Next we will try to define a cost function

```python
## We will understand this step by step
## 1) Since our y_train i.e; the labels for the price of the house is a continuous data, we will use Linear Regression to fit the data
## 2) So here our model or say function is Linear Regression
## 3) We can predict the price of the house using the function, f(x) = wx + b
## 4) Here, x is our training data
## 5) Parameters are w -> weight, b -> bias
## 6) So, if we have m training set, then the cost function is shown below
```

$$ J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} \big( f(x^{(i)}) - y^{(i)} \big)^2 $$

```python
## This is basically Mean squared error , difference to what our model predicted to that of the actual y labels
```

```python
def cost_function(X_train, y_train, w, b):
    
    m = X_train.shape[0] ## number of training set
    y_pred = w * X_train + b
    total_cost = 0
    
    for i in range(m):
        total_cost += (y_pred[i] - y_train[i])**2

    total_cost /= (2 * m)

    return total_cost
```

```python
cost_function(X_train, y_train, 135.79, 180615)
```

### So we need to find the appropriate values of w, b such that the error is minimum

```python
fig, ax = plt.subplots()
fig.suptitle('Testing Linear Regression with Random Weights and Bias', fontsize = 18)
sns.scatterplot(x = X_train, y = y_train, color = 'red', marker = '+', s = 80)
sns.lineplot(x = X_train, y = 135.79 * X_train + 180615, color = 'blue')
sns.lineplot(x = X_train, y = 100.79 * X_train + 180000, color = 'green')
sns.lineplot(x = X_train, y = 200 * X_train + 100000, color = 'orange')
ax.set_xlabel('Area of the house (sq ft)')
ax.set_ylabel('Price of the house ($)')
plt.show()
```
