```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
X = digits.data
y = digits.target
```

```python
## Checking for the KFold with example
```

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits = 3)
kf
```

```python
for train_index, test_index in kf.split([1, 2, 3, 4, 5, 6, 7, 8]):
    print(f"Train index : {train_index}, test index is : {test_index}")
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
folds = StratifiedKFold(n_splits = 3)
```

```python
from sklearn import linear_model, svm, ensemble
model_0 = linear_model.LogisticRegression(max_iter = 150)
model_1 = svm.SVC()
model_2 = ensemble.RandomForestClassifier()
```

```python
lr_score = []
svm_score = []
rf_score = []

for train_index, test_index in folds.split(X, y):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    model_0.fit(X_train, y_train)
    model_1.fit(X_train, y_train)
    model_2.fit(X_train, y_train)
    lr_score.append(model_0.score(X_test, y_test))
    svm_score.append(model_1.score(X_test, y_test))
    rf_score.append(model_2.score(X_test, y_test))
```

```python
print(f"Model score for Logistic Regression : {lr_score}")
print(f"Model score for SVM : {svm_score}")
print(f"Model score for Random Forest Classifier : {rf_score}")
```

## From the above results we see that SVM is performing better as compared to other models
